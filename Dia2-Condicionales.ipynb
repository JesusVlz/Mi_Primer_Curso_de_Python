<h3 id="comp">Operadores de comparación</h3>

Las operaciones de comparación comparan algún valor u operando y, en función de una condición, producen un valor booleano. Al comparar dos valores, puede utilizar estos operadores:

<ul>
     <li>igual: <b>==</b></li>
     <li>no igual: <b>!=</b></li>
     <li>mayor que: <b>></b></li>
     <li>menor que: <b>&lt;</b></li>
     <li>mayor o igual que: <b>>=</b></li>
     <li>menor o igual que: <b>&lt;=</b></li>
</ul>

Asignemos a <code>a</code> un valor de 5. Use el operador de igualdad denotado con dos signos iguales <b>==</b> para determinar si dos valores son iguales. El siguiente caso compara la variable <code>a</code> con 6.

```python
a=5

a==6
```

Considere el siguiente operador de comparación de igualdad: <code>i > 5</code>. Si el valor del operando izquierdo, en este caso la variable <b>i</b>, es mayor que el valor del operando derecho, en este caso 5, entonces la afirmación es <b>True</b>. De lo contrario, la afirmación es <b>False</b>. Si <b>i</b> es igual a 6, porque 6 es mayor que 5, el resultado es <b>True</b>.

```python
i=7

i>5
```

Estableciendo <codigo>i = 2</codigo>.

```python
i=2
i>5
```

Podemos aplicar los mismos métodos en cadenas. Por ejemplo, podemos usar un operador de igualdad en dos cadenas diferentes. Como las cadenas no son iguales, obtenemos un **False**.

Si usamos el operador de desigualdad, la salida será **True** ya que las cadenas no son iguales.

<h3 id="branch">Branching(Ramificaciones)</h3>


La ramificación nos permite ejecutar diferentes declaraciones para diferentes entradas. Es útil pensar en un **if** como una puerta cerrada, si la declaración es **True** podemos ingresar a la habitación y su programa ejecutará algunas tareas predefinidas, pero si la declaración es **False** el programa ignorará la tarea.

Usando lo aprendido antes en la declaración **if**. La sintaxis es tan simple como <code> if <i>instrucción</i> :</code>, que contiene una palabra <code>if</code>, cualquier instrucción y dos puntos al final. Comience sus tareas que deben ejecutarse bajo esta condición en una nueva línea con una sangría.

```python
#Por ejemplo la entrada a un bar de acuerdo a la ley
edad = 10


if edad >= 18:
  print("puede entrar a un bar")
  y=30
  print(y)

print(y)
```

La instruccion <code>else</code> ejecuta un bloque de código si ninguna de las condiciones es **True** antes del <code>else</code>.

```python
#Por ejemplo la entrada a un bar de acuerdo a la ley
edad = float(input())

print(edad)

if edad >= 18:
  print("puede entrar a un bar")

else:
  print("vete a tu casa") 
  print("no tienes acceso permito")
```

La instrucción <code>elif</code>, abreviatura de else if, nos permite verificar condiciones adicionales si las declaraciones de condición anteriores son <b>False</b>. Si la condición para la instrucción <code>elif</code> es <b>True</b>, se ejecutarán las expresiones alternativas.

```python
#Por ejemplo la entrada a un bar de acuerdo a la ley
edad = 18

if edad > 18:               #Si
  print("puede entrar a un bar")

elif edad ==18:              #Si no
  print("edad justa para entrar permitida ")

else:                         #De otra manera ejecuta esto
  print("no tiene permitido pasar")
```

 El símbolo <code> %</code> en Python se llama el Operador de Módulo. Retorna el remanente de la división del operando izquierdo por el operando derecho. Se usa para obtener el residuo de un problema de división.

*Uso* del operador módulo en la instrucción if; Hay 2 hermanas, Annie y Jane, nacidas en 1996 y 1999 respectivamente. Quieren saber quién nació en año bisiesto.

```python
Annie=1993
Jane=1996

if Annie%4==0:
    print("Annie nacio en año bisiesto")

else:
  print("Annie no nacio en año bisiesto")

if Jane%4==0:
  print("Jane nacio en año bisiesto")

else:
  print("Jane no nacio en año bisiesto")
```

Raices de una ecuacion algebraica de segundo grado de la forma:

<img src="https://www.ejerciciosecuaciones.com/wp-content/uploads/2022/08/ecuaciones-de-segundo-grado.png" width="400" align="center">

```python
a=float(input("Dame el coeficiete cuadratico"))
b=float(input("Dame el coeficiente lineal"))
c=float(input("Dame el coeficiente independiente"))

x1= (-b + (b**2-4*a*c)**(1/2))/(2*a)
x2= (-b - (b**2-4*a*c)**(1/2))/(2*a)

print("Estas son nuestras raices:", x1,x2)
```
