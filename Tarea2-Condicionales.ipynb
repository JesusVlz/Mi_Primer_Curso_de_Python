En un comedor escolar, a los niños menores de 9 años solo se les da galleta de avena y leche para el desayuno. Los niños de 10 a 14 años reciben un sandwich y los niños de 15 a 17 años una hamburguesa. El jefe de comedor pregunta la edad del alumno y le da el desayuno en consecuencia. La edad de Sam es 10. Usa la declaración if-else para determinar qué le ofrecerá el maestro de la cantina.

```python

if edad <=9:
    print ("Tendras una galleta y una lechita!")
```
